```
# Write your solution here
def distance_from_origin(a, b, c):
    return abs(c)/sqrt(a**2 + b**2)
```

```
def distance_bet_parallel(a1, b1, c1, a2, b2, c2):
    scale = a1/ a2
    c2 = c2 * scale
    return abs(c1 - c2)/sqrt(a1**2 + b1**2)
```

Write code to plot square, rectangle and ellipse

```
def sigmoid(X):
    return 1/(1 + np.exp(-X))

def tanh(X):
    return (np.exp(X) - np.exp(-X))/(np.exp(X) + np.exp(-X))

def gaussian(X):
    return np.exp(-(X**2))

def reLU(X):
    return X*(X>0)

def leakyreLU(X):
    return ((X > 0.1*X) * X) + ((X <= 0.1*X) * (0.1 * X))

def step(X):
    return X>0
```

```
X = np.linspace(-10, 10, num = 1000)
plt.plot(X, sigmoid(X))
plt.title("Sigmoid")
```

```
X = np.linspace(-10, 10, num = 1000)
plt.plot(X, tanh(X))
plt.title("tanh")
```

```
X = np.linspace(-10, 10, num = 1000)
plt.plot(X, leakyreLU(X))
plt.title("leaky relu")
```
